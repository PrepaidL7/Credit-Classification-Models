# Modelo de RegresiÃ³n LogÃ­stica - sklearn

```python
from sklearn.linear_model import LogisticRegression # Regresor Logistico
from sklearn.model_selection import train_test_split, KFold, cross_val_score, StratifiedKFold # ValidaciÃ³n, validaciÃ³n cruzada
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, me an_absolute_error # Metricas y coeficientes de eficiencia
%matplotlib inline
```

```python
# Separacion Variables predictoras y objetivo

X = practicas_pV2.drop(columns = ["status_generado", 'creditosAbiertos', 'creditosAbiertosQuebrantosNoCerrados', 'otrosCreditosParaAnalizar'])
y = practicas_pV2["status_generado"]
```

```python
# Preparacion y ajuste del modelo con todas las variables

model = LogisticRegression(solver = 'newton-cg')
model.fit(X,y)

# Se utiliza el coeficiente de determinacion R^2
model.score(X,y)
```

## Validacion del Modelo

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 1) # SepacraciÃ³n para validacion

# PrecisiÃ³n para el modelo entrenado 

name='Logistic Regression'
skfold = StratifiedKFold(n_splits=100, random_state=1, shuffle=True) 
cv_results = cross_val_score(model, X_train, y_train, cv=skfold, scoring='accuracy') # ValidaciÃ³n cruzada por dobleces estraticados
msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(msg)

# PrecisiÃ³n para las predicciones del modelo
model.fit(X_train,y_train)
preds = model.predict(X_test)
print(f'Exactitud del modelo {accuracy_score(y_test, preds)}')

y0= list(y_test)
y_pred0= list(preds)

for i in range(len(y0)):
    y0[i] = int(y0[i])
    y_pred0[i] = int(y_pred0[i])
    
from RegscorePy import aic, bic
print(f'AIC: {aic.aic(y0, y_pred0, p= len(X.columns))}')
print(f'BIC: {bic.bic(y0, y_pred0, p= len(X.columns))}')
```

```python
# Determinando el umbral de cambio para el status

both = pd.DataFrame()
both['y_pred'] = model.predict(X_train)
both['probs'] = model.predict_proba(X_train)[:,1]


for i in list(np.arange(0,1,0.1)):
    both["y_pred"] = 0 
    both.loc[both["probs"] > i, 'y_pred'] = 1
    print ("Threshold",round(i, 3),"Train Accuracy:",
    round(accuracy_score(y_train, both['y_pred']),4))
```

```python
print('Train Confusion Matrix')
print(pd.DataFrame(confusion_matrix(y_train, both['y_pred'])))

print('Test Confusion Matrix')
print(pd.DataFrame(confusion_matrix(y_test, preds)))
```

```python
print('Train Report')
print(classification_report(y_train, both['y_pred']))
print('Test Report')
print(classification_report(y_test, preds))
```

```python
from sklearn.metrics import auc, roc_curve

fpr, tpr, threshold = roc_curve(y_train, model.predict_proba(X_train)[:,1] )

roc_auc = auc(fpr, tpr)
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (area =%0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1-Specificity)')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - German Credit Data')
plt.legend(loc = 'lower right')
```

## Optimizacion del Modelo Logistico

### ClasificaciÃ³n y Retencion de Variables Predictorias

```python
# FunciÃ³n para determinar el valor de informacion de las variables

def IV_calc(data,var):
    if data[var].dtypes == "object":
        dataf = data.groupby(['bin_var'], observed=False)['status_generado'].agg(['count', 'sum'])
        dataf.columns = ["Total","bad"]
        dataf["good"] = dataf["Total"] - dataf["bad"]
        dataf["bad_per"] = dataf["bad"]/dataf["bad"].sum()
        dataf["good_per"] = dataf["good"]/dataf["good"].sum()
        dataf["I_V"] = (dataf["good_per"] - dataf["bad_per"]) * np.log(dataf["good_per"]/dataf["bad_per"])
        return dataf
    else:
        data['bin_var'] = pd.qcut(data[var].rank(method='first'),10)
        dataf = data.groupby(['bin_var'], observed=False)['status_generado'].agg(['count', 'sum'])
        dataf.columns = ["Total","bad"]
        dataf["good"] = dataf["Total"] - dataf["bad"]
        dataf["bad_per"] = dataf["bad"]/dataf["bad"].sum()
        dataf["good_per"] = dataf["good"]/dataf["good"].sum()
        dataf["I_V"] = (dataf["good_per"] - dataf["bad_per"]) * np.log(dataf["good_per"]/dataf["bad_per"])
        return dataf
```

```python
Inf_V = pd.DataFrame()
Inf_V['Variable'] = None
Inf_V['Information Value'] = None

for col in X.columns:
    IV = IV_calc(practicas_pV2, col)
    Inf_V.loc[len(Inf_V)] = [col, round(IV['I_V'].sum(), 3)]
    
print(Inf_V.sort_values('Information Value', ascending = False))    
```
